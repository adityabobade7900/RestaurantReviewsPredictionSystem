```python
foods = ["Idly", "Dosa", "Vada", "Roti", "Meals", "Veg Biryani",
         "Egg Biryani", "Chicken Biryani", "Mutton Biryani",
         "Ice Cream", "Noodles", "Manchooriya", "Orange juice",
         "Apple Juice", "Pineapple juice", "Banana juice"]


for i in foods:
    conn = pymysql.connect(user='root', password='root@7900', host='localhost',
                           database='rest_review_db')
    qur = 'INSERT INTO reviews_table(food,good_review,bad_review,customer) VALUES("%s",%d,%d,%d)' % (
        i, 0, 0, 0)
    mycur = conn.cursor()
    mycur.execute(qur)
    conn.commit()
    mycur.close()
    conn.close()
```
